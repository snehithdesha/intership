```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python


# Define the list of column names
column_names = [
    'Type', 'Alcohol', 'MalicAcid', 'Ash', 'Alcalinity',
    'Magnesium', 'Phenols', 'Flavanoids', 'Nonflavanoid',
    'Proanthocyanins', 'ColorIntensity', 'Hue', 'DilutedWines', 'Proline'
]

# Read the existing CSV file into a DataFrame
df = pd.read_csv('/content/wine.csv')

# Set the new row as the header
df.columns = column_names + list(df.columns[len(column_names):])

# Write the updated DataFrame back to the CSV file
df.to_csv('/content/wine.csv', index=False)


```

```python
wine_data=pd.read_csv("/content/wine.csv")
```

```python
X = wine_data.drop(columns='Type')
Y= wine_data['Type']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
wines=DecisionTreeClassifier(max_depth=5)
```

```python
wines.fit(X_train,Y_train)
```

```python
X_train_prediction = wines.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = wines.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```
