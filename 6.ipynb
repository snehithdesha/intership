```python
df.drop(columns=['name'] , axis=1 ,inplace= True)
df.head()
```

```python
df.head()
```

```python
df.drop(columns=['name_2'] , axis=1 ,inplace= True)
```

```python
print(df['seller_type'].unique())
```

```python
df['seller_type'] = df['seller_type'].replace({'Individual': 0, 'Dealer': 1,'Trustmark Dealer':2})
df['seller_type'] = pd.to_numeric(df['seller_type'], errors='coerce')
```

```python
print(df['transmission'].unique())
```

```python
df['transmission'] = df['transmission'].replace({'Manual': 0, 'Automatic': 1})
df['transmission'] = pd.to_numeric(df['transmission'], errors='coerce')
```

```python
print(df['owner'].unique())
```

```python
df['owner'] = df['owner'].replace({'First Owner': 0, 'Second Owner': 1,'Fourth & Above Owner':2,'Third Owner':3,'Test Drive Car':4})
df['owner'] = pd.to_numeric(df['owner'], errors='coerce')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['selling_price'], axis=1)
Y = df['selling_price']
```

```python
model=LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
training_data_prediction = model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn import metrics

score_1 = metrics.r2_score(y_train, training_data_prediction)
print("R squared error : ", score_1)
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)
print('Mean Absolute Error : ', score_2)
```
